# How to Get Specific Columns from ODS/Excel Files
Quick guide for reading your ODS file and selecting specific columns

## Setup - Import Required Modules

````python
import sys
sys.path.append('..')  # Add parent directory to path

from excel_reader import ExcelReader, quick_read_excel
import pandas as pd
````

## Method 1: Quick Read + Select Columns (Easiest)

````python
# Your file
file_path = "../alianceprofilefrom 1993to2000.ods"

# Step 1: Preview to see column names
df_preview = quick_read_excel(file_path, sheet_name=0, preview_only=True, preview_rows=4)
print("Available columns:")
print(df_preview.columns.tolist())
print("\nPreview:")
display(df_preview.head())
````

````python
# Step 2: Read full data
df = quick_read_excel(file_path, sheet_name=0)
print(f"Full data shape: {df.shape}")
````

````python
# Step 3: Select specific columns by name
# Replace 'Column1' and 'Column2' with your actual column names

# Example 1: Select by column names
# df_selected = df[['Column1', 'Column2']]

# Example 2: Select first 2 columns (if you don't know names)
column_names = df.columns.tolist()[:2]
df_selected = df[column_names]

print(f"Selected columns: {column_names}")
display(df_selected.head())
````

## Method 2: Select Columns During Read (More Efficient)

````python
# This is more efficient for large files as it only reads the columns you need

reader = ExcelReader(file_path)

# Option A: Select by column index (0-based)
df_by_index = reader.read_sheet(0, use_cols=[0, 1, 2])  # First 3 columns
print("Selected columns by index [0, 1, 2]:")
display(df_by_index.head())
````

````python
# Option B: Select by Excel-style range
df_by_range = reader.read_sheet(0, use_cols="A:C")  # Columns A to C
print("Selected columns A to C:")
display(df_by_range.head())
````

## Method 3: Using iloc (After Reading)

````python
# Read full data first
df = quick_read_excel(file_path, sheet_name=0)

# Select columns using iloc
# Get first 3 columns
df_iloc = df.iloc[:, 0:3]
print("First 3 columns using iloc:")
display(df_iloc.head())
````

````python
# Select specific column positions
df_specific = df.iloc[:, [0, 2, 4]]  # Columns at index 0, 2, and 4
print("Columns at indices [0, 2, 4]:")
display(df_specific.head())
````

## Quick Reference

```python
# By column names (after reading)
```
