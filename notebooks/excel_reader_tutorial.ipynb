# Excel Reader Tutorial
This notebook demonstrates how to use the `excel_reader` module to read and process Excel files.

## 1. Import the Module

```python
from excel_reader import ExcelReader, quick_read_excel
import pandas as pd
```

## 2. Quick Read (Simplest Method)
Use this for quick one-time reads

```python
# Replace with your Excel file path
file_path = "Rawdata/your_file.xlsx"

# Quick preview (first 5 rows)
df_preview = quick_read_excel(file_path, sheet_name=0, preview_only=True, preview_rows=5)
print("Preview:")
display(df_preview)
```

## 3. Using ExcelReader Class (Recommended)
Use this for more control and multiple operations

```python
# Create an ExcelReader instance
reader = ExcelReader(file_path)
```

### 3.1 Get File Information

```python
# Get general file info
file_info = reader.get_file_info()

print(f"File Name: {file_info['file_name']}")
print(f"File Size: {file_info['file_size_mb']} MB")
print(f"Number of Sheets: {file_info['num_sheets']}")
print(f"Sheet Names: {file_info['sheet_names']}")
```

### 3.2 Preview a Sheet

```python
# Preview first sheet (first 5 rows)
preview = reader.preview_sheet(0, rows=5)
display(preview)
```

### 3.3 Read Complete Sheet

```python
# Read entire first sheet
df = reader.read_sheet(0)

print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
display(df.head())
```

### 3.4 Get Detailed Sheet Information

```python
# Get detailed info about the sheet
sheet_info = reader.get_sheet_info(0)

print(f"Shape: {sheet_info['shape']}")
print(f"\nColumns: {sheet_info['columns']}")
print(f"\nData Types:")
for col, dtype in sheet_info['dtypes'].items():
    print(f"  {col}: {dtype}")

print(f"\nNull Counts:")
for col, count in sheet_info['null_counts'].items():
    if count > 0:
        print(f"  {col}: {count}")
```

### 3.5 Read Multiple Sheets

```python
# Read all sheets
all_sheets = reader.read_all_sheets()

# Display info for each sheet
for sheet_name, df in all_sheets.items():
    print(f"\nSheet: {sheet_name}")
    print(f"  Shape: {df.shape}")
    print(f"  Columns: {df.columns.tolist()[:5]}...")  # First 5 columns
```

### 3.6 Advanced Reading Options

```python
# Read specific columns only (first 3 columns)
df_specific_cols = reader.read_sheet(0, use_cols=[0, 1, 2])
print("Reading specific columns:")
display(df_specific_cols.head())
```

```python
# Read with skipped rows
df_skip = reader.read_sheet(0, skip_rows=2, nrows=10)
print("Reading with skipped rows:")
display(df_skip)
```

### 3.7 Export to CSV

```python
# Export first sheet to CSV
output_path = "output/exported_data.csv"
success = reader.export_sheet_to_csv(sheet_name=0, output_path=output_path, index=False)

if success:
    print(f"Successfully exported to {output_path}")
```

```python
# Search for sheets containing a keyword
keyword = "sales"  # Change this to your search term
matching_sheets = reader.search_sheets(keyword)
print(f"Sheets containing '{keyword}': {matching_sheets}")
```
